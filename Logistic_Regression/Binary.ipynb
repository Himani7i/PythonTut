```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
#create the dataset
X,y = make_classification(n_samples=1000, n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print("Accuracy Score:",score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

### GridsearchCV

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
param_grid = [
    {'penalty':['l2'], 'solver':['lbfgs','liblinear','saga','newton-cg','sag'], 'C':[0.01,0.1,1,10,100]},
    {'penalty':['l1'], 'solver':['liblinear','saga'], 'C':[0.01,0.1,1,10,100]},
    {'penalty':['elasticnet'], 'solver':['saga'], 'l1_ratio':[0.2,0.5,0.8], 'C':[0.01,0.1,1,10,100]}
]
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=param_grid,cv=cv,scoring='accuracy',n_jobs=-1)
```

```python
print(grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print("Accuracy Score:",score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
##Split yhe dataset itno train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

##split the dataset into train and test

```python
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver = ['liblinear','saga','sag','newton-cg','lbfgs']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,cv=cv,scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print("Accuracy Score:",score)  
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
