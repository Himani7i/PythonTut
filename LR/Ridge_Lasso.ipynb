```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

Data Cleaning

```python
#### missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region
1: "Bejaia Region Dataset"
2:"Sidi-BelAbbes Region Dataset"
Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset.info()
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
##Removing the null values
dataset=dataset.dropna().reset_index(drop=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
dataset=dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
###fix spaces in column names
dataset.columns=dataset.columns.str.strip()
dataset.columns
```

###Changes the required columns as integer data type


```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.info()
```

###changeing the other columns to float data type

```python
objects=[features for features in dataset.columns if dataset[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        dataset[i]=dataset[i].astype(float)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
## Lets save the cleaned datatset
dataset.to_csv('Algerian_forest_fires_cleaned.csv', index=False)
```

#### EXPLORATORY DATA ANALYSIS


```python
df=dataset.drop(['day','month','year'], axis=1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
##Encoding of the caegories in classes
df['Classes'] =np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
##Plot density plot for all features
plt.style.use('seaborn-v0_8')
```

```python
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
###Percentage for pie chart
percentage = df['Classes'].value_counts(normalize=True)*100
```

```python
##Plotting piechart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Percentage of Fire and Not Fire')
plt.show()
```

```python
df_copy=dataset
```

```python
df_copy['Classes'] =np.where(df_copy['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dftemp=df_copy.loc[df_copy['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', data=dftemp, hue='Classes')
plt.ylabel ('Number of fires', weight='bold')
plt.xlabel('Month', weight='bold')
plt.title('Monthly Fire Analysis in Sidi-bel Region ', weight='bold', size=15)
plt.show()
```

```python
dftemp=df_copy.loc[df_copy['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', data=dftemp, hue='Classes')
plt.ylabel ('Number of fires', weight='bold')
plt.xlabel('Month', weight='bold')
plt.title('Monthly Fire Analysis in Bejaia Region ', weight='bold', size=15)
plt.show()
```
