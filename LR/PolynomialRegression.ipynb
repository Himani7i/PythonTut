```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 6*np.random.rand(100,1)-3
y = 0.5*x**2 + x + 2 + np.random.randn(100,1)
#quadratic equation used = y=0.5x^2 + x + 2+outliers
plt.scatter(x,y,color='g')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#implemetn linear regression
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
regression_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,regression_1.predict(x_test))
print("R2 score for linear regression is :",score)
```

```python
#visualize this model
plt.plot(x_train,regression_1.predict(x_train),color='r')
plt.scatter(x_train,y_train,color='g')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_train_poly = poly.fit_transform(x_train)
X_test_poly = poly.transform(x_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test,y_pred)
print("R2 score for polynomial regression is :",score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(x_train,regression.predict(X_train_poly),color='b')
plt.scatter(x_train,y_train,color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
poly = PolynomialFeatures(degree=3,include_bias=True)
X_train_poly = poly.fit_transform(x_train)
X_test_poly = poly.transform(x_test)
```

```python
X_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test,y_pred)
print("R2 score for polynomial regression is :",score)
```

```python
#3 prediction of new data set
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
X_new_poly
```

```python
y_new = regression.predict(X_new_poly)
plt.plot(X_new,y_new,color='r',linewidth=2,label='new Prediction')
plt.plot(x_train,y_train,'b.',label='Training data')
plt.plot(x_test,y_test,'y.',label='Testing data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
