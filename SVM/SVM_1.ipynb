```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##data points
from sklearn.datasets import make_classification
```

```python
X,Y=make_classification(n_samples=1000,n_features=2,n_classes =2,n_clusters_per_class=1,n_redundant=0)
```

```python
X   
```

```python
Y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel ='linear')
```

```python
svc.fit(X_train,Y_train)
```

```python
##Prediction
y_pred =svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

```python
rbf = SVC(kernel ='rbf')
rbf.fit(X_train,Y_train)
```

```python
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(Y_test,y_pred1))
print(confusion_matrix(Y_test,y_pred1))
```

```python
polynomial = SVC(kernel ='poly')
polynomial.fit(X_train,Y_train)
y_pred2=polynomial.predict(X_test)
print(classification_report(Y_test,y_pred2))
print(confusion_matrix(Y_test,y_pred2))
```

```python
sigmoid = SVC(kernel ='sigmoid')
sigmoid.fit(X_train,Y_train)
y_pred3=sigmoid.predict(X_test)
print(classification_report(Y_test,y_pred3))
print(confusion_matrix(Y_test,y_pred3))
```

```python
sigmoid.intercept_
```

```python
from sklearn.model_selection import GridSearchCV
#defining parameter range
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001],'kernel': ['rbf','poly','sigmoid','linear']}   
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit = True, verbose = 3, cv=5, )
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
ygrid=grid.predict(X_test)
print(classification_report(Y_test,ygrid))
print(confusion_matrix(Y_test,ygrid))
```

## SVM Kernels indepth intuition round round data points

```python
x = np.linspace(-5.0,5.0,100)
y = np.sqrt(10**2 - x**2)
y= np.hstack([y,-y])
x= np.hstack([x,-x])
```

```python
x1 = np.linspace(-5.0,5.0,100)
y1 = np.sqrt(5**2 - x1**2)
y1= np.hstack([y1,-y1])
x1= np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
np.vstack([y,x]).T
```

```python
df1 = pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2 = pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df = pd.concat([df1,df2],axis=0)
df.head(5)
```

```python
df.tail(5)
```

```python
#Independent and dependent features
X = df.iloc[:,:2]
y = df.Y
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)  
```

Polynomial kernel
K(x,y)=(x^T * y + c)^d

```python
 #find components for the polynomial kernel
df['X1_square'] = df['X1']**2
df['X2_square'] = df['X2']**2
df['X1_X2'] = df['X1']*df['X2']
df.head(5)
```

```python
X = df[['X1','X2','X1_square','X2_square','X1_X2']]
y = df['Y']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```
